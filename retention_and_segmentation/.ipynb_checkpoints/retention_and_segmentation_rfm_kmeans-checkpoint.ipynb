## Customer Retention  

Customer Retention (удержание клиентов) — это способность компании сохранять долгосрочные отношения с клиентом, что очень важно для увеличения доходов компании. Посмотрим, как дела с удержанием покупателей у нашей компании.  
Для начала определим даты исследования

```python
print(f'Дата первой покупки {fin_df.InvoiceDate.min()}, '
      f'дата последней покупки {fin_df.InvoiceDate.max()}')
```

Таким образом, исследовать мы будем покупки за 2011 год.

```python
fin_df_11 = fin_df[pd.DatetimeIndex(fin_df.InvoiceDate).year == 2011]
```

Сделаем колонку с месяцем покупки, месяцем первой покупки (т.е. определим когорту купивших впервые в этот месяц), сгруппируем по ним и подсчитаем количество уникальных покупателей в каждый период. Далее вычислим разницу (в месяцах) между месяцем повторной и первой покупки, и развернем таблицу.

```python
fin_df_11['order_month'] = pd.DatetimeIndex(fin_df_11.InvoiceDate).to_period('M')

fin_df_11['cohort'] = pd.DatetimeIndex(fin_df_11.groupby('CustomerID_ob').
                                                InvoiceDate.transform('min')).to_period('M')
```

```python
cohort_df = fin_df_11.groupby(['cohort', 'order_month'])\
            .agg({'CustomerID_ob': 'nunique'})\
                .reset_index().rename(columns={'CustomerID_ob': 'unique_customers'})
```

```python
cohort_df['diff'] = (cohort_df.order_month - cohort_df.cohort).apply(attrgetter('n'))

cohort_pivot = cohort_df.pivot(index='cohort',columns='diff', values='unique_customers')
```

```python
cohort_pivot
```

Наконец, разделим значения ячеек на значение первой колонки, то есть вычислим долю повторных покупателей.

```python
cohort_size = cohort_pivot.iloc[:,0]
retention = cohort_pivot.divide(cohort_size, axis = 0)
retention
```

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(15, 7), sharey=True, gridspec_kw={'width_ratios': [1, 11]})

sns.heatmap(retention, 
                mask=retention.isnull(), 
                annot=True, 
                fmt='.2%', 
                cmap='BuGn',
                ax=ax[1])
ax[1].set_title('Retention\n', fontsize=16)
ax[1].set(xlabel='# of periods',
              ylabel='')
cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
white_cmap = mcolors.ListedColormap(['white'])
sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

fig.tight_layout()
```

**Вывод:** Мы видим, что около 20 процентов новых покупателей становятся нашими постоянными клиентами, и это неплохой показатель. Данные за последний месяц можно не принимать во внимание - ведь в датасете данные только за 9 дней декабря 2011 года.

## Изучение сегментации клиентов с помощью RFM-анализа и K-means 


В основе RFM-анализа лежат ключевые индикаторы поведения пользователей: дата последней покупки(recency), частота покупок(frequency) и общая сумма всех покупок(monetary).
RFM-анализ помогает ответить на следующие вопросы.
* Кто ваши самые ценные клиенты?
* На каких клиентов стоит обратить внимание, чтобы не потерять?
* Каким клиентам стоит напомнить о себе в первую очередь?


```python
fin_df_11.sample(4)
```

Сгруппироуем транзакции по дате заказа и ID покупателя, чтобы найти число заказов и сумму покупки.

```python
df_rfm = fin_df_11.groupby(['InvoiceDate', 'CustomerID_ob'])\
                    .agg({'Revenue':'sum', 'InvoiceNo': 'nunique'})\
                    .reset_index()
```

```python
df_rfm.head(3)
```

Добавим столбец с датой последнего заказа.

```python
df_rfm['last_order_time'] = df_rfm.groupby('CustomerID_ob').InvoiceDate.transform('max')
```

```python
df_rfm.head(3)
```

```python
df_rfm = df_rfm.groupby(['CustomerID_ob', 'last_order_time'])\
            .agg({'Revenue':'sum', 'InvoiceNo': 'count'})\
            .reset_index()\
            .rename(columns={'Revenue':'monetary', 'InvoiceNo': 'frequency'})
```

Определим дату отсчета и вычислим, сколько дней прошло до нее от даты последнего заказа покупателя

```python
today_date = df_rfm.last_order_time.max() + datetime.timedelta(days=1)
today_date
```

```python
df_rfm['recency'] = df_rfm.last_order_time.apply(lambda x: (today_date - x).days)
df_rfm.sort_values('frequency').tail()
```

Нам нужно разделить каждый показатель на ранги. Посмотрим на квартили наших метрик и 90 процентиль.

```python
qaintiles = df_rfm[['recency', 'frequency', 'monetary']].quantile([.25, .5, .75, .9]).to_dict()
qaintiles
```

Мы видим, что недавних покупок больше. Думаю, более наглядно будет разделить данные по recency их не по квартилям, а по кварталам календарного года.   
По frequency разделим покупателей так - 1 покупка, 2-4 покупок, 5-8 покупок, больше 8 покупок (так мы  будем знать 10 процентов  самых часто покупающих).   
По monetary - по 25, 50, 75 перцентилю (до 280, 280-615, 615-1500, 1500+)

Создадим столбцы с метриками и общий с показателем RFM.

```python
#чем меньше давность покупки тем лучше
def r_score(x):
    if x <= 90:
        return 1
    elif x <= 180:
        return 2
    elif x <= 270:
        return 3
    else:
        return 4

#чем больше денег нам принес покупатель,  тем лучше
def m_score(x):
    if x <= 280:
        return 4
    elif x <= 620:
        return 3
    elif x <= 1500:
        return 2
    else:
        return 1
#чем чаще покупают, тем лучше
def  f_score(x):
    if x == 1:
        return 4
    elif x <= 4:
        return 3
    elif x <= 8:
        return 2
    else:
        return 1
```

```python
df_rfm['R'] = df_rfm['recency'].apply(lambda x: r_score(x))
df_rfm['F'] = df_rfm['frequency'].apply(lambda x: f_score(x))
df_rfm['M'] = df_rfm['monetary'].apply(lambda x: m_score(x))
```

```python
df_rfm['RFM'] = df_rfm['R'].map(str) + df_rfm['F'].map(str)\
                        + df_rfm['M'].map(str) 
df_rfm.head()
```

Проанализируем распределение покупателей по разным метрикам. Всего сегментов - 

```python
print(*sorted(df_rfm.RFM.unique()), 'Уникальных сегментов - ', df_rfm.RFM.nunique())
```

```python
metrics = ['R', 'F', 'M']
fig, axes = plt.subplots(1, 3, figsize=(22, 5))
axes = axes.flatten()

for metric, ax in zip(metrics, axes):
    data = df_rfm.groupby(metric)\
                .agg(uniq_customers=('CustomerID_ob', 'count')).reset_index()
    sns.barplot(data=data, x='uniq_customers', y=metric, ax=ax, orient = 'h', palette="Blues_d")
    ax.set_title(f'Customers in {metric} segment')
    sns.despine()
    ax.bar_label(ax.containers[0], padding=8, fontsize=12);
```

Мы видим:

1. В метрике recency большая часть покупателей покупала что-либо в последние 3 месяца, остальные категории значительно меньше.
2. frequency - много купивших 1 или 2 раза (ранг 3 и 4), часто покупающих клиентов гораздо меньше.
3. monetary - тут без инсайтов, мы сами разделили покупателей поквартильно

Кого можно выделить и использовать в дальнейшем для различных маркетинговых нужд?

Мы убедились, что Frequency и Мonetary зависимы линейно, так что для упрощения подсчетов разделим покупателей по метрикам Recency и Frequency(предполагая что чем чаще покупают - тем больше денег приносят). Разумеется, в дальнейшем можно сегментировать и более приближенно.

Итак, наши сегменты. 

1. **Постоянные клиенты**. Покупали недавно, тратят много денег (RFM-сегмент 1Х1)
2. **Недавние клиенты**. (RFM-сегмент 1Х2, 1Х3, 1Х4)
3. **Потерянные одноразовые клиенты**. Покупали давно и денег тратили мало (RFM-сегмент 4Х3, 4Х4)
4. **Уходящие лучшие клиенты (Их очень нужно вернуть)**. Когда-то тратили очень много, но давно не покупали (RFM-сегмент 3Х1, 4Х1) 
5. **Уходящие постоянные клиенты. (Их тоже нужно вернуть)**. Как 4, но тратили меньше (RFM-сегмент 3Х2. 4Х2) 
6. **Спящие**. Тратят немного, последняя покупка - средняя по давности (RFM-сегмент [2,3]X[3,4]) 
7. **Needed attention**. Тратили больше среднего, но уже прошло некоторое время с последней покупки.(RFM-сегмент 2Х1, 2Х2) 

Добавим столбец с названием сегмента в таблицу и посмотрим на распределение клиентов по сегментам.

```python
segt_map = {
    r'11': 'regular customers',
    r'1[2-4]': 'recent customers',
    r'[3-4]1': 'very needed to retain',
    r'4[3-4]': 'lost customers',
    r'[3-4]2': 'need to retain too',
    r'[2-3][3-4]': 'sleeping',
    r'2[1-2]': 'need attention'
}

df_rfm['segment'] = df_rfm['R'].map(str) + df_rfm['M'].map(str)
df_rfm['segment'] = df_rfm['segment'].replace(segt_map, regex=True)
df_rfm['RFM_score'] = df_rfm[['R', 'F', 'M']].sum(axis=1)
df_rfm.head()
```

```python
df_rfm_segments = df_rfm.groupby('segment', as_index=False)\
            .agg(uniq_customers=('CustomerID_ob', 'count')).sort_values('uniq_customers')
df_rfm_segments.style.background_gradient(cmap=cm)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data = df_rfm_segments, y= 'segment', x='uniq_customers')
sns.despine()
ax.bar_label(ax.containers[0], padding=8);
```

## RFM clustering by K-means

Для деления на кластеры рассчитаем scores для наших метрик. Максимальный score будут иметь самые много покупающие, самые часто покупающие, и те, чья последняя покупка была совсем недавно. Минимальный - те, кто покупали мало, редко и давно. А посерединке остальные. Для этого вычислим отклонение от минимума(максимума) и нормируем на размах значений.

```python
df_rfm["F_score"]= ((df_rfm['frequency'] - df_rfm['frequency'].min()) 
                    / (df_rfm['frequency'].max() - df_rfm['frequency'].min())) 
df_rfm["M_score"]= ((df_rfm['monetary'] - df_rfm['monetary'].min()) 
                    / (df_rfm['monetary'].max() - df_rfm['monetary'].min())) 
df_rfm["R_score"]= ((df_rfm['recency'].max()- df_rfm['recency']) / 
                    (df_rfm['recency'].max() - df_rfm['recency'].min()))
```

Найдем оотптимальное число кластеров методом локтя (Elbow-Method using within-cluster sum of squares)

```python
WCSS = []
for i in range(1,11):
    model = KMeans(n_clusters = i,init = 'k-means++')
    model.fit(df_rfm[['R_score','F_score','M_score']])
    WCSS.append(model.inertia_)
fig = plt.figure(figsize = (6,4))
plt.plot(range(1,11),WCSS, linewidth=4, markersize=12,marker='o',color='red')
plt.xticks(np.arange(11))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
sns.despine()
plt.show()
```

Определим число кластеров как 4

```python
kmeans = KMeans(4)
```

Разделим на кластеры и посмотрим  на распределение покупателей в них

```python
identified_clusters = kmeans.fit_predict(df_rfm[['R_score','F_score','M_score']])
df_rfm['Clusters'] = identified_clusters
df_rfm['Clusters'].value_counts()
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
for m, zlow, zhigh in [('o', 1, 1), ('o', 1, 1)]:
    xs = df_rfm.F_score
    ys = df_rfm.R_score
    zs = df_rfm.M_score
    ax.scatter(xs, ys, zs, marker=m,c=df_rfm["Clusters"], s=40, cmap="gist_rainbow")
ax.set_xlabel('Frequency score -->')
ax.set_ylabel('Recency score -->')
ax.set_zlabel('Monetary score -->')
```
